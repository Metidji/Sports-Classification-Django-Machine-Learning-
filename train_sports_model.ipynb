In this lab we will train a simple model for task of Sports Classificiation using the publically available dataset: https://www.kaggle.com/datasets/gpiosenka/sports-classification

```python
import kagglehub

path = kagglehub.dataset_download('gpiosenka/sports-classification')
```

```python
import tensorflow as tf
import os

train_data = tf.keras.utils.image_dataset_from_directory(
    os.path.join(path, 'train'),
    labels="inferred",
    label_mode="int",
    color_mode="rgb",
    batch_size=32,
    image_size=(224, 224),
    shuffle=True,
)

valid_data = tf.keras.utils.image_dataset_from_directory(
    os.path.join(path, 'valid'),
    labels='inferred',
    label_mode='int',
    color_mode='rgb',
    batch_size=32,
    image_size=(224, 224),
    shuffle=True,
)
```

```python
backbone = tf.keras.applications.MobileNetV3Small(
        input_shape=(224, 224, 3),
        include_top=False,
        weights='imagenet',
        pooling='avg',
        include_preprocessing=True,
    )

backbone.trainable = False

f = 16
model = tf.keras.models.Sequential()
model.add(backbone)
model.add(tf.keras.layers.Dense(100))
model.summary()
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
```

```python
model.fit(train_data,
          validation_data=valid_data,
          epochs=100)
```

```python
model.save('sports.keras')
```
