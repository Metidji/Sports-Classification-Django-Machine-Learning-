```python
import numpy as np
import cv2

img = cv2.cvtColor(cv2.imread('test-img.jpg'), cv2.COLOR_BGR2RGB)

img_int8 = (img.astype(np.float32) - 128).astype(np.int8)

# 3. Save as .h file
with open("test_img.h", "w") as f:
    f.write("const int8_t test_image_data[] = {\n")
    # Flatten the 224x224x3 array into a long list of numbers
    flat_data = img_int8.flatten()
    for i, val in enumerate(flat_data):
        f.write(str(val) + (", " if i < len(flat_data) - 1 else ""))
        if i % 12 == 0: f.write("\n") # Just for readability
    f.write("\n};")

print("test_image.h generated!")
```
